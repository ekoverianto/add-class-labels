```
f_open = open('sample_data/AppleStore.csv') #select repository for your data

from csv import reader
f_read = reader(f_open)
data = list(f_read)

#if the price of the app is 0.0, print 'free', else print 'buy'
for val in data[1:]:
  price = float(val[5])
  if price == 0.0:
    val.append('Free')
  else:
    val.append('Buy')
    
data[0].append('Free Or Buy')

#converting list of lists into a table
import pandas as pd
pd.DataFrame(data)
```
